##### 0) 라이브러리 불러오기

```python
# (1) 자연어처리 라이브러리
import gensim

# (2) 딥러닝 라이브러리
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

# (3) 기타 라이브러리
import numpy as np
```

```python
torch.manual_seed(1)
```

##### 1) Pytorch Word Embedding 흐름

(1) Word2Idx 정의 !

```python
word_to_idx = {"hello": 0, "world": 1}
```

(2) 모델(Word_Embedding_Matrix) 정의

```python
embedding = nn.Embedding(2, 5)

embedding
```

(3) LookUp Tensor 추출 !!

```python
lookup_tensor = torch.tensor([word_to_idx["hello"]], dtype=torch.long)

lookup_tensor
```

(4) 모델(Word_Embedding_Matrix) 결과 출력

```python
embedding(lookup_tensor)
```

##### 4) Pretrained 모델(Word Embedding Matrix) 불러오기 + Unkown Word2Vec 추가

```python
word2vec_model.add_vector("<unk>", np.zeros(50)) # !!

Word_Embedding_Matrix = torch.FloatTensor(word2vec_model.vectors)
```

##### 5) 모델(Word_Embedding_Matrix) 정의

```python
embedding = nn.Embedding.from_pretrained(Word_Embedding_Matrix) # !
embedding.requires_grad = False # !!
# True : 학습하는 도중, NLI든 Sentiment Analysis든 Word Embeddding도 함께 학습됨 -> 속도 느려짐
# False : 학습하는 도중, Word Embedding 함께 학습되지 않음 -> 속도 빨라짐
```
